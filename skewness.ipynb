```python
import pandas as pd
from datatransform import DataTransform
from dataframeinfo import DataFrameInfo
from dataframetransform import DataFrameTransform
from plotter import Plotter

# creates the dataframe from the .csv downloaded from AWS Relational Database
df = pd.read_csv('loan_payments_null_imputations.csv')

data = DataTransform(df)
df_info = DataFrameInfo(df)
plotter = Plotter(df)
df_transform = DataFrameTransform(df)
```

```python
skewed_columns_list = df_info.identify_skewed_columns(threshold=0.5)

for column in skewed_columns_list:
    transformed_log_column = plotter.log(column)

for column in skewed_columns_list:
    transformed_boxcox = plotter.boxcox(column)

for column in skewed_columns_list:
    transformed_yeojohnson = plotter.yeojohnson(column)






```
